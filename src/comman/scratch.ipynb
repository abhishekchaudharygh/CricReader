```python
import pandas as pd
import matplotlib.pyplot as plt

matches = pd.read_csv('/Users/abhishek.chaudhary/Abhishek/Developer/CricReader/src/dataset/matches.csv')
deliveries = pd.read_csv('/Users/abhishek.chaudhary/Abhishek/Developer/CricReader/src/dataset/deliveries.csv')
```

```python
team = 'Mumbai Indians'

team_b_df = deliveries[deliveries['batting_team'] == team]
team_b_batsman_runs_sum = team_b_df.groupby('batter')['batsman_runs'].sum().reset_index()
team_b_batsman_runs_sum.rename(columns={'batsman_runs': 'total_batsman_runs'}, inplace=True)
team_b_batsman_runs_sum = team_b_batsman_runs_sum.sort_values(by='total_batsman_runs', ascending=False)
top_n_run_scorers = team_b_batsman_runs_sum.head(5)
plt.bar(top_n_run_scorers['batter'], top_n_run_scorers['total_batsman_runs'])
plt.xlabel('Batter')
plt.ylabel('Runs Scored')
plt.title(f'Top run scorers for {team}')
plt.show()
```

```python
team = 'Royal Challengers Bangalore'

team_b_df = deliveries[deliveries['bowling_team'] == team]
team_b_bowler_wicket_sum = team_b_df.groupby('bowler')['is_wicket'].sum().reset_index()
team_b_bowler_wicket_sum.rename(columns={'is_wicket': 'total_bowler_wickets'}, inplace=True)
team_b_bowler_wicket_sum = team_b_bowler_wicket_sum.sort_values(by='total_bowler_wickets', ascending=False)
top_n_wicket_taker = team_b_bowler_wicket_sum.head(5)
plt.bar(top_n_wicket_taker['bowler'], top_n_wicket_taker['total_bowler_wickets'])
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.title(f'Top wicket taker for {team}')
plt.show()
```
